#Exploratory data analysis

#Country with most attacks

```python
# Total count of Target type in the DataFrame
df['country_txt'].value_counts()
```

```python
print("Countries with the most attacks:",df['country_txt'].value_counts().index)
```

```python
plt.subplots(figsize=(20, 6))
sns.barplot(df['country_txt'].value_counts().index[:40],df['country_txt'].value_counts()[:40].values,palette='magma')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.title('Country with the most attacks', fontsize =20)
plt.show()
```

```python
print("Country with the most attacks:",df['country_txt'].value_counts().idxmax())
```

#Region by most attack

```python
# Total count of Region type in the DataFrame
df['region_txt'].value_counts()
```

```python
print("Region with the most attacks:",df['region_txt'].value_counts().index)
```

#Attack Year VS Region

```python
dff_region=pd.crosstab(df.iyear,df.region)
dff_region.plot(color=sns.color_palette('Set2',12))
fig=plt.gcf()
plt.title("Terrorist Attack Year VS Region", fontsize=20)
fig.set_size_inches(20,7)
plt.show()
```

#Attack Year VS Target Type

```python
dff_region=pd.crosstab(df.iyear, df.targtype1)
dff_region.plot(color=sns.color_palette('Set2',12))
fig=plt.gcf()
plt.title("Terrorist Attack Year VS Target Type", fontsize=20)
fig.set_size_inches(15,8)
plt.show()
```
